```python
class parent:
    def method(self):
        print("parent method")
class child(parent):
    def method(self):
        print("child method")
obj = child()
obj.method()
```

```python
from multipledispatch import dispatch

class Test:
    @dispatch(int, int)
    def add(self, a, b):
        print(a + b)

    @dispatch(int, int, int)
    def add(self, a, b, c):
        print(a + b + c)

    @dispatch(str, str)
    def add(self, a, b):
        print(a + b)

obj = Test()
obj.add(1, 2)
obj.add(1, 2, 3)
obj.add("Hello ", "World")
```
