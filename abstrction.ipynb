```python
from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def method(self):
        pass
    def method2(self):
        print("conc")
    @abstractmethod
    def method3(self):
        pass
    @abstractmethod
    def method4(self):
        pass
        
class B(A):
    def method(self):
        print("abstract")
    def method3(self):
        print("abstract")
    def method4(self):
        print("abstract")
obj = B()
obj.method()
obj.method2()
obj.method3()
```
