```python
class Account:
    def __init__(self,name,balance,pin):
        self.name = name
        self._balance = balance
        self.__pin = pin
    def get_pin(self):
        return self.__pin
    def set_pin(self,new_pin):
        self.__pin = new_pin
    def show_balance(self):
        return self._balance

acc = Account("nanai",1000,1234)
print(acc.name)
print(acc._balance)
print(acc.get_pin())
print(acc.get_pin())
acc.set_pin(123)
print(acc.get_pin())
acc.show_balance()
```
